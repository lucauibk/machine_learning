# Power Consumption Prediction — Team Overfitters
**Test Notebook** — loads the saved model and evaluates on the test set.

> Do NOT retrain here. This notebook only loads the model and predicts.

```python
import numpy as np
import pandas as pd
import joblib
from sklearn.metrics import mean_squared_error, r2_score

MODEL_PATH = 'models/final_model.pkl'
DATA_PATH  = '../instructions_examples/Datasets/powerpredict.csv'

print('Loading model...')
model = joblib.load(MODEL_PATH)
print('Model loaded.')
```

```python
# Load test data
# On JupyterHub the test CSV is provided separately; adjust DATA_PATH if needed.
df_test = pd.read_csv(DATA_PATH)

# Drop weather_description (same preprocessing as training)
desc_cols = [c for c in df_test.columns if 'weather_description' in c]
X_test = df_test.drop(columns=['power_consumption'] + desc_cols, errors='ignore')

# If ground truth is available, report metrics
if 'power_consumption' in df_test.columns:
    y_test = df_test['power_consumption'].values
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2   = r2_score(y_test, y_pred)
    print(f'Test RMSE: {rmse:.2f} MW')
    print(f'Test R²:   {r2:.4f}')
else:
    y_pred = model.predict(X_test)
    print(f'Predictions generated for {len(y_pred)} samples.')

print('\nFirst 10 predictions:')
print(np.round(y_pred[:10], 1))
```
